```python
# Now suppose we need to find the dactorial of a number (or for drama of 1000 numbers).
```

```python
# One way will be to create a fuction using loops.
def our_fact(n):
    fact = 1
    for i in range(1, n+1):
        fact *= i
    return fact
```

```python
factorial_of_5 = our_fact(5)
```

```python
factorial_of_5
```

```python
# Good, now let's look at the easier way to do that.
# For this we will need the functools library and operator.
```

```python
from functools import reduce
from operator import mul
```

```python
# functools stores a lot of very usefull functions like reduce, which we will use now
# operator stored a lot of callble functions for pair operators like +, -, > etc.
```

```python
factorial_of_5 = reduce(mul, range(1, 6))
```

```python
factorial_of_5
```

```python
# The same result. Now let's time them.
```

```python
%%timeit
factorial_of_5 = our_fact(5)
```

```python
%%timeit
factorial_of_5 = reduce(mul, range(1, 6))
```

```python
# And here we are, we gained a small, but speed up.
```

```python
# map function.
```

```python
# map function allows us to apply a function on iterable data structure, usually a list
```

```python
# Let's look at a example.
# In linear algebra every thing is a vector (actually a tensor, by for a 2d space it is a vector)
```

```python
# Vectors have a property - we can add and multiply a scalar to them, meaning we add or 
# multiply every element of the vector with this scalar.
```

```python
# Let's create a test vector
vector = [4, 5, 6, 7]
```

```python
# Now let's add to it a scalar.
```

```python
res = map(lambda x : x + 3, vector)
```

```python
res
```

```python
# It is returning a mapping object, we must convert it into a list
```

```python
list(res)
```

```python
# as you see we added 3 to every element of the vector/list
```

```python
# Now let's multiply our list/vector with a scalar
```

```python
res = map(lambda x : x * 3, vector)
```

```python
list(res)
```

```python
# And we are done.
```

```python
# filter function.
# The filter function allows us to get all values that make as function return true.
```

```python
# The function that we apply as a filter must return True or False.
```

```python
# For example let's extract from a list of random values to extract all even numbers.
```

```python
import random
random.seed(42)
rand_list = [random.randint(0, 10000) for i in range(10_000)]
```

```python
# Let's look at our list
```

```python
rand_list
```

```python
# Good now let's filter this list.
even_value = filter(lambda x : x % 2 == 0, rand_list)
```

```python
even_value
```

```python
# Also we get a filter object.
# We must convert it into a list
```

```python
list(even_value)
```
