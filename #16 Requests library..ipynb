```python
# For this lecture we will get data from spotify.
# Firstly we must create an account if you still doesn't have one.
# And make it an developer account.
```

```python
# Now Let's import the requsts library.
import requests
```

```python
# Now we need to setup the CLIENT_ID and the CLIENT_SECRET
CLIENT_ID = '1e8df497a8254e4b8ee14e464f164d71'
CLIENT_SECRET = '652617f9f29740229e0a58f7f63951a8'
```

```python
# To get the data from spotify we must firstly autetithicate ourselves, let's define the 
# Autehtification url
AUTH_URL = 'https://accounts.spotify.com/api/token'
```

```python
# Now we can send the aiutetification request
auth_response = requests.post(AUTH_URL, {
    'grant_type': 'client_credentials',
    'client_id': CLIENT_ID,
    'client_secret': CLIENT_SECRET,
})
```

```python
# Now we must use the json funtion to convert the response into json format
auth_response_data = auth_response.json()
```

```python
# Now we can get the accese token to get the data
access_token = auth_response_data['access_token']
```

```python
# Let's extract the headers
headers = {
    'Authorization': 'Bearer {token}'.format(token=access_token)
}
```

```python
# Now we can finaly request data from spotify
# base URL of all Spotify API endpoints
BASE_URL = 'https://api.spotify.com/v1/'

# Track ID from the URI
track_id = '6y0igZArWVi6Iz0rj35c1Y'

# actual GET request with proper header
r = requests.get(BASE_URL + 'audio-features/' + track_id, headers=headers)
```

```python
# Now let's look at the data from the response
r.json()
```

```python
# Suppose we now need to get all the albums of the a certain artist
```

```python
# Firstly we must get the artis id. We will use the Led's Zeppelin id
```

```python
artist_id = '36QJpDe2go2KgaRleHCDTp'
```

```python
# We must request the data from a specific link that has the next format
# 'https://api.spotify.com/v1/artists/{artist_id}/albums'
```

```python
# Now we can send the request
r = requests.get(BASE_URL + 'artists/' + artist_id + '/albums', 
                 headers=headers)
```

```python
r.json()
```

```python
# We got a preatty big json.
# Now we must clean it.
```

```python
data = r.json()['items']
```

```python
# Let's look at our data
data
```

```python
# It's a list of values, let's see how many elements does is has
len(data)
```

```python
# However every album has a diferent number of tracks, let's see how many do we have
track_num = 0
for album in data:
    track_num += album['total_tracks']
track_num
```

```python
# We have 303 tracks, now let's try to get them.
```

```python
album_ids = [album['id'] for album in data]
```

```python
album_ids
```

```python
# Now we can use these ids to get all the tracks
```

```python
r = requests.get(BASE_URL + 'albums/' + '6VH2op0GKIl3WNTbZmmcmI' + '/tracks', 
                headers=headers)
```

```python
r
```

```python
# Goot it's a good request
```

```python
# Now lt's look at the data inside
r.json()
```

```python
# Now using the list of album_id let's get the names of all tracks
tracks = []
for album_id in album_ids:
    r = requests.get(BASE_URL + 'albums/' + album_id + '/tracks', 
                headers=headers)
    data = r.json()['items']
    for track in data:
        tracks.append(track['name'])
```

```python
# Let's look how many tracks did we get.
len(tracks)
```
